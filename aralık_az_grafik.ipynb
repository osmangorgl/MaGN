```python
import pandas as pd
import matplotlib.pyplot as plt

# Dosyayı yükle
data = pd.read_excel('/content/1.xlsx')

# Ölçüm verilerini çiz
plt.scatter(data.index, data['Sağ'], label='Sağ', color='blue')
plt.scatter(data.index, data['SOL'], label='Sol', color='red')

# Üst sınır, alt sınır ve nominal değer çizgilerini ekle
plt.axhline(67, label='Üst Sınır', color='green', linestyle='--')
plt.axhline(47, label='Alt Sınır', color='orange', linestyle='--')
plt.axhline(57, label='Nominal', color='purple')

# Y eksenindeki özel değerleri göster
current_yticks = list(plt.yticks()[0])
additional_yticks = [47, 57, 67]
plt.yticks(sorted(set(current_yticks + additional_yticks)))



# Etiketler ve başlık ekle
plt.xlabel('Ölçüm İndeksi')
plt.ylabel('Ölçüm Değeri (mm)')
plt.title('1. Fikstür Ölçüm Değerleri')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dosya yolu Colab'da '/content/' ile başlar
data_path = '/content/.xlsx'

# Veri dosyasını yükle
data = pd.read_excel(data_path)

# 'Sağ' ve 'SOL' sütunlarını düşen verileri çıkartarak birleştir
right = data['Sağ'].dropna()
left = data['SOL'].dropna()

# Grafik boyutunu belirle
plt.figure(figsize=(14, 7))

# Sağ ve sol ölçümleri çiz
plt.scatter(right.index, right, color='blue', label='Sağ (Right)')
plt.scatter(left.index, left, color='red', label='Sol (Left)')

# Üst sınır, alt sınır ve nominal değeri çizgi olarak ekle
plt.axhline(67, color='green', linestyle='--', label='Üst Sınır (Upper Limit)')
plt.axhline(47, color='orange', linestyle='--', label='Alt Sınır (Lower Limit)')
plt.axhline(57, color='purple', linestyle='-', label='Nominal')

# x eksenindeki tüm indeksleri göster
plt.xticks(data.index)

# Başlık ve etiketleri ekle
plt.title('1. Fikstür Ölçüm Değerleri')
plt.xlabel('Ölçüm İndeksi')
plt.ylabel('Ölçüm Değeri (mm)')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Grafiği göster
plt.show()
```
